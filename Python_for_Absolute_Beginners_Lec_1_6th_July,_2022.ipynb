# Program Structure - Ways to structure a program

There are three program structures of Python Programming.
1.   Sequential Steps
2.   Conditional Steps
3.   Repeated Steps



**Sequential Steps:**
The program flows from one step to another. This is called Sequential Programming.

```python
x = 10 = int(input("Enter the Value:"))

y = x * 3

print(y)
```

**Conditional Steps:** The flow of the program depends on the conditions.

```python
#line1
x = int(input("Enter your Value:")) #ex

#Conditional Structure
if x < 5:

  print("lower")

elif x > 5:

  print("bigger.")

#line3
print("close")
```

**Repeated Steps:**

---
The program will run again and again. These repeated steps are called loops in Python


```python
y = range(-5)
print(y)
```

```python
x = 5
for i in range(x):
  x -= 1  # x = x - 1 # x +=1 ==> x = x + 1
  print(x)
print("Close")
```

# Elements in Python

**Variables**

---
Words in which the values can be stored and used later on are known as variables

```python
x = int(input("Enter your 1st number:"))
y = int(input("Enter your 2nd number:"))
z = x + y
if z < 10:
  print("The sum is: ",z)
  print("lower")
elif z > 10:
  print("The sum is: ",z)
  print("higher")
```
