**Assignment Statements in Python**

---
Using assignment operator, we can assign the value to the variable


```python
#Assignement Statment
x = 30

# y = x + 2  # This is assignment with expression. x  + 2 = y
# x + 2 = y
print(y)  #Built-in print fucntion to print out the output.
```

*   We assign the value to the variable using an assignement sign "**=**".
*   An **assignment statement** consists of an expression on the right hand side of "**=**" and a **variable** to store the result.

```python
#For Example

x = x + 10

Name  = "I am Muhammad Talha."

x = 5.1 * (x/10)**2   #Equation
```
